### Used Selenium with javascript which was succesful to extract the elements from the website


```python
base_url = "https://www.nasdaq.com/market-activity/stocks/"


company_symbols = ["AAPL"]


driver = webdriver.Chrome()


for symbol in company_symbols:
    
    
    company_url = f"{base_url}{symbol}"+"/financials"

 
    driver.get(company_url)
    
    panel_names = ["balanceSheetTable","incomeStatementTable"]

   
    table_data_dict = {}

    for panel_name in panel_names:
    

        script = f'''
            var table = document.querySelector('div[data-panel-name="{panel_name}"] table.financials__table');
            var data = [];
            if (table) {{
                var rows = table.querySelectorAll('tbody tr');
                var column_headers = table.querySelectorAll('thead tr th[scope="col"]');
                var header_data = Array.from(column_headers, function (header) {{ return header.textContent.trim(); }});
                data.push(header_data);

                rows.forEach(function (row) {{
                    var row_cells = row.querySelectorAll('th[scope="row"], td');
                    var row_data = Array.from(row_cells, function (cell) {{ return cell.textContent.trim(); }});
                    data.push(row_data);
                }});
            }}
            return data;
        '''



        result = driver.execute_script(script)

   
        table_data_dict[panel_name] = result




driver.quit()
```
